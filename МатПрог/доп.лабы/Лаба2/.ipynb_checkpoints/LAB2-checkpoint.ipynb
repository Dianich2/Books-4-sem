```python
nq = logs.platform.nunique()
nq
```

```python
logs.query('success == True').platform.value_counts().idxmax()
```

```python
logs.query('success == True').groupby('client').agg({'success':'sum'}).sort_values('success', ascending=False).query('success == 41')
```

```python
merge_data = logs.merge(user_data, on='client')
merge_data
```

```python
merge_data.query('premium == True').groupby('platform').agg(count=('platform', 'count')).sort_values('count', ascending = False)
```

```python
premium_clients = merge_data.query('premium == True')
premium_clients
```

```python
unpremium_clients = merge_data.query('premium == False')
unpremium_clients
```

```python
a = sns.countplot(x = 'platform', data=premium_clients)
a.set(xlabel='Platform', ylabel='Count')
sns.despine()
plt.show()
```

```python
b = sns.countplot(x = 'platform', data=unpremium_clients)
b.set(xlabel='Platform', ylabel='Count')
sns.despine()
plt.show()
```

```python
c = sns.countplot(data=merge_data.loc[(merge_data.platform == 'computer') & (merge_data.success == True)], x='age')
c.set(xlabel='Age', ylabel='CountSuccess')
c.set_xticklabels(c.get_xticklabels(), rotation=90)
sns.despine()
plt.figure(figsize=(12,8))
plt.show()
```

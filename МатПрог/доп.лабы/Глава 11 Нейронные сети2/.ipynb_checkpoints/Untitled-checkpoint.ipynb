```python
import tensorflow as tf 
import sklearn as skl
import numpy as np
from tensorflow import keras
```

```python
tf.random.set_seed(42)
np.random.seed(42)
```

```python
# СҒРҫР·РҙР°РҪРёРө РјРҫРҙРөР»Рё СҒ 20 СҒР»РҫСҸРјРё Dense
my_model = keras.models.Sequential()
my_model.add(keras.layers.Flatten(input_shape=[32, 32, 3]))
for i in range(20):
    my_model.add(keras.layers.Dense(100, activation="elu", kernel_initializer="he_normal"))
my_model.add(keras.layers.Dense(10, activation="softmax"))

optimizer = keras.optimizers.Nadam(learning_rate=5e-5)
my_model.compile(loss="sparse_categorical_crossentropy", metrics=["accuracy"], optimizer=optimizer)
```

```python
# Р·Р°РіСҖСғР·РәР° РҙР°СӮР°СҒРөСӮР°
(x_train_all, y_train_all), (x_test_all, y_test_all) = keras.datasets.cifar10.load_data()
```

```python
# СҖР°Р·РұРёРІР°РөРј СӮСҖРөРҪРёСҖРҫРІРҫСҮРҪСӢР№ РҪР°РұРҫСҖ РҪР° СӮСҖРөРҪРёСҖРҫРІРҫСҮРҪСӢР№ Рё РҝСҖРҫРІРөСҖРҫСҮРҪСӢР№
x_train = x_train_all[5000:]
y_train = y_train_all[5000:]

x_valid = x_train_all[:5000]
y_valid = y_train_all[:5000]
```

```python
# РҝРөСҖРөРјРөРҪРҪСӢРө РҙР»СҸ callbacks
early_stopping = keras.callbacks.EarlyStopping(patience=10)
checkpoints = keras.callbacks.ModelCheckpoint("my_model.h5", save_best_only=True)
all_callbacks = [early_stopping, checkpoints]
```

```python
my_model.fit(x_train, y_train, epochs=1000, validation_data=(x_valid, y_valid), callbacks=all_callbacks) 
```

```python
my_best_model = keras.models.load_model("my_model.h5")
my_best_model.evaluate(x_valid, y_valid)
```

```python
# Р—РҗР”РҗРқРҳР• РЎ
keras.backend.clear_session()
tf.random.set_seed(42)
np.random.seed(42)
model_with_BN = keras.models.Sequential()
model_with_BN.add(keras.layers.Flatten(input_shape=[32,32,3]))
model_with_BN.add(keras.layers.BatchNormalization())
for i in range(20):
    model_with_BN.add(keras.layers.Dense(100, kernel_initializer="he_normal"))
    model_with_BN.add(keras.layers.BatchNormalization())
    model_with_BN.add(keras.layers.Activation("elu"))
model_with_BN.add(keras.layers.Dense(10, activation="softmax"))
```

```python
optimizer = keras.optimizers.Nadam(learning_rate=5e-4)
model_with_BN.compile(loss="sparse_categorical_crossentropy", optimizer=optimizer, metrics=["accuracy"])

# РҝРөСҖРөРјРөРҪРҪСӢРө РҙР»СҸ callbacks
early_stopping = keras.callbacks.EarlyStopping(patience=10)
checkpoints = keras.callbacks.ModelCheckpoint("my_model5.h5", save_best_only=True)
all_callbacks = [early_stopping, checkpoints]
```

```python
model_with_BN.fit(x_train, y_train, epochs=1000, validation_data=(x_valid, y_valid), callbacks=all_callbacks) 
```

```python
my_best_model = keras.models.load_model("my_model5.h5")
my_best_model.evaluate(x_valid, y_valid)
```
